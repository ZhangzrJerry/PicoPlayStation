### 一、随机大法
让机器随机地进行游戏操作

```python
from random import random
import matplotlib.pyplot as plt


gamer_ = [Numbergame() for i in range(500)]
point_ = {i:0 for i in range(gamer.length())}
for gamer in gamer_:
    while True:
        if gamer.if_win():
            break
        if gamer.if_end():
            break
        gamer.random_create()
        # 随机选择上下左右一种走法
        randnum = random()
        if randnum < 0.25:
            gamer.up()
        elif randnum < 0.50:
            gamer.down()
        elif randnum < 0.75:
            gamer.left()
        else:
            gamer.right()
        print(gamer.show())
    point_[max(max(gamer.show()))] += 1
plt.plot(point_.values())
plt.show()
```
